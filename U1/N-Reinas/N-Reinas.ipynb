Ejemplo de tablero de N-reinas: <br>

[0,6,4,7,1,3,5,2] <-- Ejemplo de solución para 8 reinas

Explicación: <br>
La posición en la lista representa la columna del tablero, y el dígito <br>
que está en esa posición representa la fila en que la reina está colocada. <br>
En este caso, la primera reina está en la esquina superior izquierda.

Ya que en este problema tenemos que validar cualquier número de reinas, se generará <br>
un tablero N x N que contendrá N reinas.

```python
# Verifica si hay colisiones entre reinas en el tablero.
def verificar_colisiones(tablero):
    for i in range(len(tablero)):
        #print("i = {}".format(i))
        for j in range(i + 1, len(tablero)):
            #print("j = {}".format(j))
            if tablero[i] == tablero[j] or abs(tablero[i] - tablero[j]) == abs(i - j):
                #print("entró")
                return True
    #print("fuera")
    return False
```

```python
# Genera nuevas configuraciones de reinas a partir del tablero actual.
def generar_soluciones(tablero, n):
    soluciones = []
    col = len(tablero)  # Columna actual a considerar

    # Solo procesamos si aún no hemos colocado todas las reinas
    if col < n:
        for fila in range(n):
            nueva_config = tablero[:]  # Copiamos el tablero actual
            nueva_config.append(fila)  # Añadimos la nueva fila en la columna actual
            #print("fila: {}".format(nueva_config))
            if not verificar_colisiones(nueva_config):
                print("Sin colision: {}".format(nueva_config))
                soluciones.append(nueva_config)
    
    return soluciones
```

```python
# Función para imprimir el tablero
def imprimir_tablero(solucion):
    n = len(solucion)
    for fila in range(n):
        for col in range(n):
            if solucion[col] == fila:
                print("X", end=" ")
            else:
                print(".", end=" ")
        print()  # Nueva línea después de cada fila
```

```python
#Implementación del algoritmo de búsqueda BFS
def BFS(n):
    frontera = [[]]  # Comienza con un tablero vacío
    while frontera:
        cabeza = frontera.pop(0)

        # Si hemos colocado todas las reinas
        if len(cabeza) == n:
            return cabeza

        # Generar nuevas configuraciones de reinas
        nuevas_soluciones = generar_soluciones(cabeza, n)
        frontera.extend(nuevas_soluciones)

    return None  # No se encontró solución
```

```python
def play_bfs():
    n = 8  # Número de reinas
    solBFS = BFS(n)
    
    if solBFS is not None:
        print("Solución encontrada: {}".format(solBFS))
        #print("Cantidad recorrida con BFS: {}".format(len(solBFS)))
        imprimir_tablero(solBFS)
    else:
        print("No se encontró solución")
```

```python
play_bfs()
```

```python
#Implementación del algoritmo de búsqueda DFS
def DFS(n):
    frontera = [[]]  # Comienza con un tablero vacío
    while frontera:
        cabeza = frontera.pop() # Simular pila en lugar de cola

        # Si hemos colocado todas las reinas
        if len(cabeza) == n:
            return cabeza

        # Generar nuevas configuraciones de reinas
        nuevas_soluciones = generar_soluciones(cabeza, n)
        frontera.extend(nuevas_soluciones)

    return None  # No se encontró solución
```

```python
def play_dfs():
    n = 8  # Número de reinas
    solDFS = DFS(n)
    
    if solDFS is not None:
        print("Solución encontrada: {}".format(solDFS))
        #print("Cantidad recorrida con DFS: {}".format(len(solDFS)))
        imprimir_tablero(solDFS)
    else:
        print("No se encontró solución")
```

```python
# Genera nuevas configuraciones de reinas a partir del tablero actual.
def generar_soluciones_limitado(tablero, n, limite):
    soluciones = []
    col = len(tablero)  # Columna actual a considerar

    # Solo procesamos si aún no hemos colocado todas las reinas
    if col < n:
        for fila in range(n):
            nueva_config = tablero[:]  # Copiamos el tablero actual
            nueva_config.append(fila)  # Añadimos la nueva fila en la columna actual
            #print("fila: {}".format(nueva_config))
            if not verificar_colisiones(nueva_config) and fila <= limite:
                print("Sin colision: {}".format(nueva_config))
                soluciones.append(nueva_config)
    
    return soluciones
```

```python
#Implementación del algoritmo de búsqueda LDFS
def LDFS(n, limite):
    frontera = [[]]  # Comienza con un tablero vacío
    while frontera:
        cabeza = frontera.pop() # Simular pila en lugar de cola

        # Si hemos colocado todas las reinas
        if len(cabeza) == n:
            return cabeza

        # Generar nuevas configuraciones de reinas
        nuevas_soluciones = generar_soluciones_limitado(cabeza, n, limite)
        frontera.extend(nuevas_soluciones)

    return None  # No se encontró solución
```

```python
def play_ldfs():
    n = 8  # Número de reinas
    limite = 5
    solLDFS = LDFS(n, limite)
    
    if solLDFS is not None:
        print("Solución encontrada: {}".format(solLDFS))
        #print("Cantidad recorrida con DFS: {}".format(len(solDFS)))
        imprimir_tablero(solLDFS)
    else:
        print("No se encontró solución")
```

```python
play_ldfs()
```

```python
#Implementación del algoritmo de búsqueda ILDFS
def ILDFS(n, limite_inicial):
    limite_actual = limite_inicial

    while True:
        #Si el limite es mayor a uno establecido
        if limite_actual >= 100:
            #entonces no encontro una solición (Evita enciclarse por siempre)
            print("No se encontro solución dentro del limite")
            return None
        sol = LDFS(n, limite_actual)

        if sol is not None:
            print("Se encontro la solución en el nivel: {}".format(limite_actual+1))
            return sol

        limite_actual += 1
        print("No se encontro solución en el nivel: {}".format(limite_actual))
```

```python
def play_ildfs():
    n = 8  # Número de reinas
    limite = 0
    solILDFS = ILDFS(n, limite)
    imprimir_tablero(solILDFS)
    if len(solILDFS) == 0:
        print("No se encontró solución")
        return
```

```python
def heuristica(ladoIzq):
    m, c, _ = ladoIzq
    return m + c
```

```python
def evaluacionVoraz(posibilidades, visitados):
    best_val = 100000
    best = None
    for posible in posibilidades:
        if posible not in visitados:
            val = heuristica(posible[0])
            if val < best_val:
                best = posible
                best_val = val
    return best
```

```python
def eliminaPosibilidadeIguales(posibilidades, lista):
  nueva_lista = []
  for posible in posibilidades:
    if posible in lista:
      continue
    nueva_lista.append(posible)
  return nueva_lista
```

```python
def voraz(nodo, visitados, solucion, meta, n_iteraciones):
  n_iteraciones = n_iteraciones + 1
  if nodo == None:
    return []
  print("Recorriendo, actualmente en: {}".format(nodo))
  solucion.append(nodo)
  if criterioAceptacion(nodo, meta):
    print("Se ha llegado a la meta: {}".format(nodo))
    print("Número de iteraciones: {}".format(n_iteraciones))
    return solucion
  print("nodo: {}".format(nodo))
  nuevas_soluciones = generarSoluciones(nodo[0], nodo[1])
  print("Nuevas soluciones: {}".format(nuevas_soluciones))
  #print(nuevas_soluciones)
  nuevas_soluciones = eliminaPosibilidadeIguales(nuevas_soluciones, visitados)
  #print(nuevas_soluciones)
  mejor = evaluacionVoraz(nuevas_soluciones, visitados)
  #print(mejor)
  visitados.append(nodo)
  return voraz(mejor, visitados, solucion, meta, n_iteraciones)
```

```python
def play_voraz():
  inicio = ([0,0,0], [3,3,1])
  meta = ([3,3,1], [0,0,0])
  visitados, solucion = [], []
  solVoraz = voraz(inicio, visitados, solucion, meta, 0)
  print("Cantidad recorrida con voraz: {}".format(len(solVoraz)))
  # El recorrido no se imprimirá si no llega a la solución, pero se puede ver mientras recorre nodos.
  print("Recorrido: {}".format(solVoraz))
  imprimir_tablero(solVoraz)
  if len(solVoraz) == 0:
    print("No se encontro solucion")
    return
```

```python
#Función que evalua cada posibilidad y ordena la lista de menor a mayor en este caso.
def evaluacion_aStar(posibilidades, visitados):
  evaluaciones = []
  for posible in posibilidades:
    if posible not in visitados:
        #Se suma g(n) + h(n)
        valor = posible[1] + heuristica(posible[0][0])
        evaluaciones.append([posible, valor])
  #print(evaluaciones)
  def second_value(elemento):
    return elemento[1]
  evaluaciones.sort(key= second_value)
  #print(evaluaciones)
  resultado = [posible for (posible, valor_evaluacion) in evaluaciones]
  #print(resultado)
  return resultado
```

```python
def AStar(frontera_act, visitados, solucion, meta, n_iteraciones):
  n_iteraciones = n_iteraciones + 1
  frontera = frontera_act[:]
  print("Inicio",frontera)
  if len(frontera) == 0:
    return []
  nodo = frontera.pop(0)
  solucion.append(nodo[0])
  if criterioAceptacion(nodo[0], meta):
    return solucion
  print("Nod", nodo[0])
  nuevas_soluciones = generarSoluciones(nodo[0][0], nodo[0][1])
  print("Sol",nuevas_soluciones)
  nuevas_soluciones = eliminaPosibilidadeIguales(nuevas_soluciones, visitados)
  print("No obs", nuevas_soluciones)
  nuevas_soluciones_costo = []
  for sol in nuevas_soluciones:
    nuevas_soluciones_costo.append((sol, nodo[1] + 1))
  frontera = frontera + nuevas_soluciones_costo
  #print(frontera)
  frontera_ordenada = evaluacion_aStar(frontera, visitados)
  visitados.append(nodo[0])
  #LLamado recursivo
  return AStar(frontera_ordenada, visitados, solucion, meta, n_iteraciones)
```

```python
def play_AStar():
  inicio = ([0,0,0], [3,3,1])
  meta = ([3,3,1], [0,0,0])
  inicio = (inicio, 0)
  visitados, solucion = [], []
  solAStar = AStar([inicio], visitados, solucion, meta, 0)
  print("Cantidad recorrida con AStar: {}".format(len(solAStar)))
  # El recorrido no se imprimirá si no llega a la solución, pero se puede ver mientras recorre nodos.
  print("Recorrido: {}".format(solAStar))
  imprimir_tablero(solAStar)
  if len(solAStar) == 0:
    print("No se encontro solucion")
    return
```
